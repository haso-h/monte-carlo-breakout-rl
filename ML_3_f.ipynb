```python

import time, random
from collections import defaultdict
from dataclasses import dataclass
from typing import List, Tuple, Dict
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from tqdm import trange

# ────────────────────────────────────────────────
# Environment
# ────────────────────────────────────────────────
ACTIONS = [-1, 0, 1]

@dataclass
class BreakoutEnv:
    width: int = 15
    height: int = 10
    brick_width: int = 3
    paddle_width: int = 5
    max_paddle_speed: int = 2
    brick_rows: int = 2
    brick_cols: int = 4
    layout_type: str = "rectangle"
    seed: int = 0

    def __post_init__(self):
        random.seed(self.seed)
        self.np_random = np.random.RandomState(self.seed)
        self.brick_positions = self._init_brick_positions()
        self.reset()

    def _init_brick_positions(self) -> List[Tuple[int, int]]:
        bricks = []
        for r in range(self.brick_rows):
            cols_row = min(self.brick_cols, r + 1) if self.layout_type == "triangle" else self.brick_cols
            margin_x = (self.width - cols_row * self.brick_width) // 2
            for c in range(cols_row):
                x = margin_x + c * self.brick_width
                if self.layout_type == "zigzag":
                    x += (r % 2) * (self.brick_width // 2)
                bricks.append((x, 1 + r))
        return bricks

    def reset(self):
        self.paddle_x = (self.width - self.paddle_width) // 2
        self.paddle_v = 0
        self.bricks = set(self.brick_positions)
        self.ball_x = self.width // 2
        self.ball_y = self.height - 2
        self.ball_vx, self.ball_vy = random.choice([-2, -1, 0, 1, 2]), -1
        return self._get_state()

    def _get_state(self):
        return (
            self.ball_x, self.ball_y, self.ball_vx, self.ball_vy,
            self.paddle_x, self.paddle_v, tuple(sorted(self.bricks))
        )

    def _paddle_range(self):
        return range(self.paddle_x, self.paddle_x + self.paddle_width)

    def step(self, action: int):
        assert action in ACTIONS
        self.paddle_v = int(np.clip(self.paddle_v + action, -self.max_paddle_speed, self.max_paddle_speed))
        self.paddle_x = int(np.clip(self.paddle_x + self.paddle_v, 0, self.width - self.paddle_width))

        self.ball_x += self.ball_vx
        self.ball_y += self.ball_vy

        # Wall collision
        if self.ball_x < 0:
            self.ball_x, self.ball_vx = -self.ball_x, -self.ball_vx
        if self.ball_x >= self.width:
            self.ball_x, self.ball_vx = 2*(self.width-1)-self.ball_x, -self.ball_vx
        if self.ball_y < 0:
            self.ball_y, self.ball_vy = -self.ball_y, -self.ball_vy

        reward = -1  # Step penalty

        # Paddle collision
        if self.ball_y == self.height - 1:
            if self.ball_x in self._paddle_range():
                hit = self.ball_x - self.paddle_x
                self.ball_vx = hit - (self.paddle_width // 2)
                self.ball_vy = -1
                reward = 1  # Paddle hit reward
            else:
                return self._get_state(), -1, True, {"result": "missed"}

        # Brick collision
        hit = None
        for bx, by in list(self.bricks):
            if by == self.ball_y and bx <= self.ball_x < bx + self.brick_width:
                self.bricks.remove((bx, by))
                self.ball_vy *= -1
                reward += 10
                break

        done = len(self.bricks) == 0
        if done:
            reward += 100

        return self._get_state(), reward, done, {"result": "win" if done else None}

# ────────────────────────────────────────────────
# Monte Carlo Control
# ────────────────────────────────────────────────
def mc_control(
        env: BreakoutEnv,
        episodes=70000,
        gamma=0.99,
        epsilon_start=1.0,
        epsilon_end=0.02,
        success_target=0.90,
        window=2000,
        log_every=2000,
        min_episodes=10000
    ):
    Q, ret_sum, ret_cnt = defaultdict(float), defaultdict(float), defaultdict(int)
    policy = {}
    R_hist, S_hist, L_hist, W_hist, F_hist = [], [], [], [], []
    early_ep = episodes

    for ep in trange(episodes, desc="Training", ncols=80):
        ε = epsilon_start + (epsilon_end - epsilon_start) * min(1.0, ep / episodes)
        state = env.reset()
        traj, steps, done, G = [], 0, False, 0.0

        while not done:
            action = random.choice(ACTIONS) if (random.random() < ε or state not in policy) else policy[state]
            next_state, reward, done, info = env.step(action)
            traj.append((state, action, reward))
            G += reward
            state = next_state
            steps += 1

        win = info.get("result") == "win"
        fail = info.get("result") == "missed"
        R_hist.append(G)
        S_hist.append(int(win))
        L_hist.append(steps)
        W_hist.append(win)
        F_hist.append(fail)

        # Logging and early stopping
        if (ep + 1) % log_every == 0:
            avg_succ = np.mean(S_hist[-log_every:]) * 100
            avg_len = np.mean(L_hist[-log_every:])
            print(f"Ep {ep+1}/{episodes} | ε={ε:.2f} | Avg win% {avg_succ:.1f} | "
                  f"Avg len {avg_len:.1f}")

        # First-visit MC update
        G_ret, visited = 0.0, set()
        for s, a, r in reversed(traj):
            G_ret = gamma * G_ret + r
            if (s, a) not in visited:
                visited.add((s, a))
                ret_sum[(s, a)] += G_ret
                ret_cnt[(s, a)] += 1
                Q[(s, a)] = ret_sum[(s, a)] / ret_cnt[(s, a)]
                policy[s] = max(ACTIONS, key=lambda aa: Q[(s, aa)])

        if ep >= min_episodes and ep > window and np.mean(S_hist[-window:]) >= success_target:
            print(f"✓ Early stop at episode {ep+1} – moving win rate ≥ {success_target*100:.0f}%")
            early_ep = ep + 1
            break

    return policy, R_hist, S_hist, L_hist, W_hist, F_hist, early_ep

# ────────────────────────────────────────────────
# Visualization and Reporting
# ────────────────────────────────────────────────
def moving_avg(arr, w=500):
    arr = np.array(arr)
    if len(arr) < w:
        return arr
    return np.convolve(arr, np.ones(w) / w, mode="valid")

def plot_curve(values, title, ylabel="Value"):
    plt.figure(figsize=(6, 3))
    plt.plot(values, alpha=0.5, label='Per episode')
    plt.plot(moving_avg(values), label=f"Moving avg")
    plt.title(title)
    plt.xlabel("Episode")
    plt.ylabel(ylabel)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()

def plot_trajectory(env: BreakoutEnv, policy: Dict, title="", show_paddle=True):
    state = env.reset(); xs, ys = [env.ball_x], [env.ball_y]; paddle_xs = [env.paddle_x]
    for _ in range(300):
        action = policy.get(state, random.choice(ACTIONS))
        state, _, done, _ = env.step(action)
        xs.append(env.ball_x); ys.append(env.ball_y); paddle_xs.append(env.paddle_x)
        if done: break

    fig, ax = plt.subplots(figsize=(7, 4))
    ax.set_xlim(0, env.width); ax.set_ylim(env.height, 0); ax.set_aspect('equal')
    ax.grid(alpha=0.3)
    ax.plot(xs, ys, marker='o', color='blue', label='Ball')
    if show_paddle:
        ax.plot(paddle_xs, [env.height-1]*len(paddle_xs), color='grey', marker='_', label='Paddle')
    for bx, by in env.brick_positions:
        ax.add_patch(patches.Rectangle((bx, by), env.brick_width, 1, fill=True, alpha=0.25, edgecolor='black'))
    ax.set_title(f"Trajectory – {title}")
    ax.legend()
    plt.tight_layout()
    plt.show()

def plot_multiple_trajectories(env: BreakoutEnv, policy: Dict, n=3, title=""):
    for i in range(n):
        plot_trajectory(env, policy, title=f"{title} (Trial {i+1})", show_paddle=(i==0))

def stats_summary(name, L_hist, S_hist, W_hist, F_hist, ep90, wall_time):
    arr = np.array(L_hist)
    print(f"\n=== {name} RESULTS ===")
    print(f"Episodes run: {ep90}")
    print(f"Win rate:      {100*np.mean(S_hist):.1f}%")
    print(f"Loss rate:     {100*np.mean(F_hist):.1f}%")
    print(f"Median length: {np.median(arr):.1f}")
    print(f"Mean length:   {np.mean(arr):.1f}")
    print(f"Min/Max len:   {np.min(arr)}, {np.max(arr)}")
    print(f"Wall time:     {wall_time:.1f} s")

# ────────────────────────────────────────────────
# Layout configurations
# ────────────────────────────────────────────────
LAYOUTS = [
    dict(name="Rectangle 2×4", brick_rows=2, brick_cols=4, layout_type="rectangle"),
    dict(name="Zigzag 3×3",    brick_rows=3, brick_cols=3, layout_type="zigzag"),
    dict(name="Triangle 3",    brick_rows=3, brick_cols=3, layout_type="triangle"),
]

# ────────────────────────────────────────────────
# Main loop
# ────────────────────────────────────────────────
def run_experiments():
    for cfg in LAYOUTS:
        env = BreakoutEnv(**{k: v for k, v in cfg.items() if k != "name"}, seed=42)
        print(f"\n=== {cfg['name']} | bricks = {len(env.brick_positions)} ===")
        t0 = time.perf_counter()
        policy, R, S, L, W, F, ep90 = mc_control(env)
        wall_time = time.perf_counter() - t0

        stats_summary(cfg['name'], L, S, W, F, ep90, wall_time)
        plot_curve(R, f"{cfg['name']} – Return per episode", ylabel="Total return")
        plot_curve(np.array(S)*100, f"{cfg['name']} – Win %", ylabel="Win %")
        plot_curve(L, f"{cfg['name']} – Episode length", ylabel="Episode length")
        plot_multiple_trajectories(env, policy, n=3, title=cfg['name'])

if __name__ == "__main__":
    run_experiments()
```
